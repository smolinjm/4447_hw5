```python
import seaborn
titanic_df = seaborn.load_dataset('titanic')
titanic_df.head(4)
```

```python
p_titanic = titanic_df.drop_duplicates('age').pivot(index='age',columns='class',values='fare')
p_titanic.head()
```

```python
p_titanic = titanic_df.pivot_table(index='age',columns='class',values='fare')
p_titanic.head()
```

```python
print('AGGREGATE BY MEAN/MIN...')
p_titanic = titanic_df.pivot_table(index='age',columns='class',values='fare', aggfunc=np.min)
p_titanic.head()
```

```python
titanic_copy = p_titanic.copy()
titanic_copy.columns = titanic_copy.columns.astype(str)
melted_titanic_df = pd.melt(
    titanic_copy.reset_index(),
    id_vars='age',
    var_name='class_renamed',
    value_vars=['First','Second','Third'],
    value_name='fare'
)
print(len(titanic_copy))
melted_titanic_df.sample(5)
```
